```python
# yaml 파일을 학습이 가능하도록 경로 설정.

import yaml

data = { 'train' : '/content/Aquarium_Data/train/images/',
        'val' : '/content/Aquarium_Data/valid/images/',
         'test' : '/content/Aquarium_Data/test/images',
         'names' : ['fish', 'jellyfish', 'penguin', 'puffin', 'shark', 'starfish', 'stingray'],
         'nc': 7 }

with open('/content/Aquarium_Data/Aquarium_Data.yaml', 'w') as f:  # 새로운 yaml파일이름설정 Aquarium_Data.yaml
  yaml.dump(data, f) #위 colab링크경로로 변경한 data를 사용해 Aquarium_Data.yaml생

with open('/content/Aquarium_Data/Aquarium_Data.yaml', 'r') as f: # 잘만들어졌는지 확인위해 화면출력
  aquarium_yaml = yaml.safe_load(f)
  display(aquarium_yaml)
```

```python
import ultralytics

ultralytics.checks()
```

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt') # MS COCO Dataset으로 사전학습된 YOLOV8N 모델로드
```

```python
print(type(model.names), len(model.names))

print(model.names)
```

```python
# YOLOv8 로 커스텀 데이터학습하기** 반드시 train함수의 파라미터로 앞 수정된 yaml화일로 지정
model.train(data='/content/Aquarium_Data/Aquarium_Data.yaml', epochs=100, patience=30, batch=32, imgsz=416)
```

```python
print(type(model.names), len(model.names))

print(model.names)

# Aquarium_Data.yaml의 커스텀데이터로 학습됐기 때문에, 학습 후 model.names값을 보면,
#  사전학습된 MS COCO데이터의 80개가 아닌 새로 yaml파일에서 설정한 7개의 클래스와 이름으로 바뀌어 있음
```

```python
result = model.predict(source = '/content/Aquarium_Data/test/images/', save=True)
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
import os
# runs/detect/predict2 예측이미지
runurl = 'runs/detect/predict2'
runurllist = os.listdir(runurl) # 변경된이미지들을 리스트화

predict_array_list = []
for i in range(len(runurllist)):
  loaded_img = load_img(os.path.join(runurl, runurllist[i]), target_size=(100, 100))
  #loaded_img = load_img(runurllist[i], target_size=(300, 300))
  loaded_img_array = img_to_array(loaded_img) / 255.0

  predict_array_list.append(loaded_img_array)

plt.figure(figsize = (100, 100))

# for i in range(len(img_array_list)): range() 함수는 함수의 전달인자에 기반하여 정수 순열을 제공합니다
# for i in range(len(predict_array_list[10])):  # 10번째 이미지까지만 출력되게
for i in range(10):  # 10번째 이미지까지만 출력되게
  #plt.subplot(1, len(predict_array_list[10]), i+1)
  plt.subplot(2, 5, i+1) # 2개씩해서 5줄은?
  #plt.subplot(총 행 개수, 총 열 개수, 그래프 번호) https://heytech.tistory.com/448
  plt.xticks([]); plt.yticks([])
  plt.title(predict_array_list[i]) # 목록 인덱스가 범위를 벗어남??
  plt.imshow(predict_array_list[i])
  # 500은 자꾸 다운됨
```
